```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/notmaddie3932/pompeii-graffiti/main/abbreviation_data.csv")
```

```python
df.sample(10)
```

### Abbreviations by Locations

```python
num_per_loc = alt.Chart(df).mark_bar().encode(
    y = "location:N",
    x = "count():Q"
).properties(
    title = "Number of Abbreviations by Location"
)

num_per_loc
```

### Abbreviations by Abbreviation Type and Word Type

```python
alt.Chart(df).mark_bar().encode(
    x='word type:N',
    y='count():Q',
    color='abbreviation type:N'
).properties(
    title = 'Distribution of Abbreviation Types by Word Types',
    width = 500,
    height = 500
)
```

```python
# get domestic oonly
df_domestic = df[df['location'] == 'domestic']
domestic_word_counts = df_domestic['word type'].value_counts(normalize=True) * 100
df_domestic_counts = pd.DataFrame({'word type': domestic_word_counts.index, 'value':domestic_word_counts.values}) # df with word and %

df_lupanare = df[df['location'] == 'lupanare']
lupanare_word_counts = df_lupanare['word type'].value_counts(normalize=True) * 100
df_lupanare_counts = pd.DataFrame({'word type': lupanare_word_counts.index, 'value':lupanare_word_counts.values}) # df with word and %
```

```python
domestic = alt.Chart(df_domestic_counts).mark_arc().encode(
    theta='value:Q',
    color=alt.Color('word type:N', scale=alt.Scale(scheme='tableau10'))
).properties(
    title = 'domestic',
)

lupanare = alt.Chart(df_lupanare_counts).mark_arc().encode(
    theta='value:Q',
    color=alt.Color('word type:N', scale=alt.Scale(scheme='tableau10'))
).properties(
    title='lupanare',
)

full = domestic|lupanare

full.properties(title='Word Type by Location')
```
